```python
vendors_by_state = df['state'].value_counts()
print("Number of vendors by state:")
print(vendors_by_state)


vendors_by_region = df['region'].value_counts()
print("\nDistribution of vendors across different regions:")
print(vendors_by_region)

unique_contacts_per_vendor = df.groupby('vendor_name')['contact_person'].nunique()
print("\nUnique contact persons per vendor:")
print(unique_contacts_per_vendor)
```

```python
average_rating = df.groupby('product_category')['rating'].mean().reset_index()

# Question 2: How does warranty time vary across different product categories?
warranty_time_variation =  df[['product_category', 'warranty_time']]

# Question 3: Which product categories are offered by the most vendors?
product_category_count = df['product_category'].value_counts().reset_index()
product_category_count.columns = ['product_category', 'count']

# Visualizations
plt.figure(figsize=(14, 10))

# Visualization 1: Average Rating for Each Product Category
plt.subplot(2, 2, 1)
sns.barplot(y='rating', x='product_category', data=average_rating, palette='viridis')
plt.title('Average Rating for Each Product Category')
plt.ylabel('Average Rating')
plt.xlabel('Product Category')

# Visualization 2: Warranty Time Variation Across Product Categories
plt.subplot(2, 2, 2)
sns.boxplot(x='warranty_time', y='product_category', data=warranty_time_variation, palette='viridis')
plt.title('Warranty Time Variation Across Product Categories')
plt.xlabel('Warranty Time (Years)')
plt.ylabel('Product Category')

# Visualization 3: Product Categories Offered by the Most Vendors
plt.subplot(2, 2, 3)
sns.barplot(x='count', y='product_category', data=product_category_count, palette='viridis')
plt.title('Product Categories Offered by the Most Vendors')
plt.xlabel('Number of Vendors')
plt.ylabel('Product Category')

plt.tight_layout()
plt.show()
```

```python
# Question 1: What is the average delivery time for each region?
average_delivery_time = df.groupby('region')['delivery_time'].mean().reset_index()

# Question 2: How do transportation charges vary by state?
transportation_charges_variation = df[['state', 'transportation_charges']]

# Question 3: Is there any correlation between delivery time and transportation charges?
correlation_data = df[['delivery_time', 'transportation_charges']]

# Visualizations
plt.figure(figsize=(16, 12))

# Visualization 1: Average Delivery Time for Each Region
plt.subplot(2, 2, 1)
sns.barplot(x='delivery_time', y='region', data=average_delivery_time, palette='coolwarm')
plt.title('Average Delivery Time for Each Region')
plt.xlabel('Average Delivery Time (Days)')
plt.ylabel('Region')

# Visualization 2: Transportation Charges Variation by State
plt.subplot(2, 2, 2)
sns.boxplot(x='transportation_charges', y='state', data=transportation_charges_variation, palette='coolwarm')
plt.title('Transportation Charges Variation by State')
plt.xlabel('Transportation Charges (Currency)')
plt.ylabel('State')



plt.tight_layout()
plt.show()
```

```python
top_vendors_by_sales = df.groupby('vendor_name')['sales'].sum().reset_index().sort_values(by='sales', ascending=False)

# Question 2: What is the relationship between quantity sold and profit?
quantity_vs_profit = df[['quantity', 'profit']]

# Question 3: How do discounts impact the profit margins?
discount_vs_profit = df[['discount', 'profit']]

# Visualizations
plt.figure(figsize=(18, 12))

# Visualization 1: Vendors with the Highest Sales Figures
plt.subplot(2, 2, 1)
sns.barplot(x='sales', y='vendor_name', data=top_vendors_by_sales.head(10), palette='viridis')
plt.title('Top 10 Vendors by Sales Figures')
plt.xlabel('Sales (Currency)')
plt.ylabel('Vendor Name')
```

```python
df['delivery_date'] = pd.to_datetime(df['delivery_date'])
sales_over_time = df.groupby('delivery_date')['sales'].sum().reset_index()

plt.figure(figsize=(14, 7))
sns.lineplot(x='delivery_date', y='sales', data=sales_over_time)
plt.title('Line Graph of Sales Over Time')
plt.xlabel('Delivery Date')
plt.ylabel('Sales (Currency)')
plt.show()
```

```python


# Question 1: Vendors with the Highest and Lowest Average Ratings
vendor_ratings = df.groupby('vendor_name').agg({
    'rating': 'mean',
    'profit': 'mean',
    'sales': 'sum'
}).reset_index()

highest_rated_vendors = vendor_ratings.sort_values(by='rating', ascending=False).head(5)
lowest_rated_vendors = vendor_ratings.sort_values(by='rating').head(5)

# Visualization: Highest and Lowest Rated Vendors with Profit Margins and Sales
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
sns.barplot(x='rating', y='vendor_name', data=highest_rated_vendors, palette='viridis')
plt.title('Top 10 Highest Rated Vendors')
plt.xlabel('Average Rating')
plt.ylabel('Vendor Name')

plt.subplot(1, 2, 2)
sns.barplot(x='rating', y='vendor_name', data=lowest_rated_vendors, palette='viridis')
plt.title('Top 10 Lowest Rated Vendors')
plt.xlabel('Average Rating')
plt.ylabel('Vendor Name')

plt.tight_layout()
plt.show()


```

```python


# Calculate total sales and average profit margins for each product category
category_sales_profit = df.groupby('product_category').agg({
    'sales': 'sum',
    'profit': 'mean',
    'quantity': 'sum'
}).reset_index()

# Visualization: Highest Sales and Profit Margins by Product Category
plt.figure(figsize=(14, 7))

# Sales by Product Category
plt.subplot(1, 2, 1)
sns.barplot(x='sales', y='product_category', data=category_sales_profit, palette='viridis')
plt.title('Total Sales by Product Category')
plt.xlabel('Total Sales (Currency)')
plt.ylabel('Product Category')

# Profit Margins by Product Category
plt.subplot(1, 2, 2)
sns.barplot(x='profit', y='product_category', data=category_sales_profit, palette='viridis')
plt.title('Average Profit Margins by Product Category')
plt.xlabel('Average Profit Margin (Currency)')
plt.ylabel('Product Category')

plt.tight_layout()
plt.show()

# Visualization: Effect of Quantity Sold on Sales and Profit
```

```python


# Calculate total sales for each region
sales_by_region = df.groupby('region').agg({
    'sales': 'sum',
    'transportation_charges': 'mean',
    'delivery_time': 'mean'
}).reset_index()

# Identify regions with the highest and lowest sales
highest_sales_region = sales_by_region.loc[sales_by_region['sales'].idxmax()]
lowest_sales_region = sales_by_region.loc[sales_by_region['sales'].idxmin()]

# Visualization: Sales, Transportation Charges, and Delivery Time by Region
plt.figure(figsize=(18, 6))

# Total Sales by Region
plt.subplot(1, 3, 1)
sns.barplot(x='region', y='sales', data=sales_by_region, palette='viridis')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales (Currency)')

# Transportation Charges by Region
plt.subplot(1, 3, 2)
sns.barplot(x='region', y='transportation_charges', data=sales_by_region, palette='viridis')
plt.title('Average Transportation Charges by Region')
plt.xlabel('Region')
plt.ylabel('Average Transportation Charges (Currency)')

# Delivery Time by Region
plt.subplot(1, 3, 3)
sns.barplot(x='region', y='delivery_time', data=sales_by_region, palette='viridis')
plt.title('Average Delivery Time by Region')
plt.xlabel('Region')
plt.ylabel('Average Delivery Time (Days)')

plt.tight_layout()
plt.show()
```

```python

# Extract month and season from delivery dates
df['delivery_date'] = pd.to_datetime(df['delivery_date'])
df['month'] = df['delivery_date'].dt.month
df['season'] = df['delivery_date'].dt.month % 12 // 3 + 1
season_dict = {1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'}
df['season'] = df['season'].map(season_dict)

# Calculate monthly and seasonal sales, profit, and delivery times
monthly_data = df.groupby('month').agg({
    'sales': 'sum',
    'profit': 'sum',
    'delivery_time': 'mean'
}).reset_index()

seasonal_data = df.groupby('season').agg({
    'sales': 'sum',
    'profit': 'sum',
    'delivery_time': 'mean'
}).reset_index()

# Visualization: Monthly Sales, Profit, and Delivery Times
plt.figure(figsize=(18, 6))

# Monthly Sales
plt.subplot(1, 3, 1)
sns.lineplot(x='month', y='sales', data=monthly_data, marker='o', palette='viridis')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales (Currency)')

# Monthly Profit
plt.subplot(1, 3, 2)
sns.lineplot(x='month', y='profit', data=monthly_data, marker='o', palette='viridis')
plt.title('Monthly Profit')
plt.xlabel('Month')
plt.ylabel('Total Profit (Currency)')

# Monthly Delivery Time
plt.subplot(1, 3, 3)
sns.lineplot(x='month', y='delivery_time', data=monthly_data, marker='o', palette='viridis')
plt.title('Monthly Delivery Time')
plt.xlabel('Month')
plt.ylabel('Average Delivery Time (Days)')

plt.tight_layout()
plt.show()


```

```python
# Track the ratings of vendors over time
vendor_ratings_over_time = df.groupby(['delivery_date', 'vendor_name'])['rating'].mean().reset_index()

# Calculate the trend in ratings for each vendor
vendor_rating_trends = vendor_ratings_over_time.groupby('vendor_name').apply(lambda x: x.set_index('delivery_date')['rating'].rolling(window=12, min_periods=1).mean()).reset_index()

# Identify vendors with consistent improvement or decline
improving_vendors = vendor_rating_trends.groupby('vendor_name')['rating'].apply(lambda x: x.diff().mean()).reset_index()
improving_vendors.columns = ['vendor_name', 'rating_trend']
improving_vendors = improving_vendors.sort_values(by='rating_trend', ascending=False)

# Highlight top 5 improving and declining vendors
top_improving_vendors = improving_vendors.head(5)
top_declining_vendors = improving_vendors.tail(5)

# Visualization: Top Improving Vendors
plt.figure(figsize=(14, 7))
for vendor in top_improving_vendors['vendor_name']:
    vendor_data_subset = vendor_rating_trends[vendor_rating_trends['vendor_name'] == vendor]
    sns.lineplot(x='delivery_date', y='rating', data=vendor_data_subset, label=vendor)
plt.title('Top 5 Improving Vendors\' Rating Trends')
plt.xlabel('Date')
plt.ylabel('Rating')
plt.legend()
plt.show()

# Visualization: Top Declining Vendors
plt.figure(figsize=(14, 7))
for vendor in top_declining_vendors['vendor_name']:
    vendor_data_subset = vendor_rating_trends[vendor_rating_trends['vendor_name'] == vendor]
    sns.lineplot(x='delivery_date', y='rating', data=vendor_data_subset, label=vendor)
plt.title('Top 5 Declining Vendors\' Rating Trends')
plt.xlabel('Date')
plt.ylabel('Rating')
plt.legend()
plt.show()
```

```python
# Calculate average profit margins for different warranty times
warranty_data = df.groupby('warranty_time').agg({
    'profit': 'mean'
}).reset_index()

# Visualization: Profit Margins by Warranty Time
plt.figure(figsize=(14, 7))
sns.barplot(x='warranty_time', y='profit', data=warranty_data, palette='viridis')
plt.title('Profit Margins by Warranty Time')
plt.xlabel('Warranty Time (Months)')
plt.ylabel('Average Profit Margin (Currency)')
plt.show()

# Statistical Analysis: Check if there's a significant difference in profit margins based on warranty time
from scipy.stats import f_oneway

# Group profit margins by warranty time
warranty_groups = [group['profit'].values for name, group in df.groupby('warranty_time')]

# Perform ANOVA test
anova_result = f_oneway(*warranty_groups)
anova_result
```
